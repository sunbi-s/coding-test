```python
'''
for ~ esle 가 틀리는 바람에 한참을 해맷다. 왜틀렸지??

for문이 전부 돌면 else 그 외에는 break가 발동한다.
시간 효율 올리려고 2단 break 때매 쓴건데 음..

**********************************추가 ***************************************
 flag를 places for문 밖에 선언해서 틀렸다!
나머지 29개가 맞은게 기적이다. 어이가 없네;
이거때매 거의 1시간 반을 날렸다...
'''
def check(row, col, table, step):
    #종료조건
    step += 1
    if step == 3:
        return 1
    flag = 1
    moves = [[1,0],[0,1],[0,-1]]
    # 현재 위치 제거
    table[row][col] = 'X'
    #1칸 이동
    
    for move in moves:
        # 중간 종료
        if flag == 0:
            return 0
        n_row = row + move[0]
        n_col = col + move[1]
        # 맵밖으로 나감
        if n_row > 4 or n_row < 0 or n_col > 4 or n_col < 0:
            continue
        # p있으면 바로 아웃
        if table[n_row][n_col] == 'P':
            return 0
        # O 인경우 추가탐색
        if table[n_row][n_col] == 'O':
            flag = check(n_row, n_col, table, step)
        # X 인 경우 아무것도 안함
    
    return flag
            

def solution(places):
    answer = []
    # 여기다가 flag =1 선언해서 틀림;
    
    for place in places:
        flag = 1
        table = []
        for i in range(len(place)):
            table.append(list(place[i])) #Casting   
        
        for row in range(5):
            for col in range(5):
                if table[row][col] == 'P': #탐색시작
                    flag = check(row, col, table, 0)
                    if flag == 0:
                        break
            else:
                continue
            break
        answer.append(flag)
    return answer
```

```python
'''
다른 사람의 풀이 
재귀가 아닌 while 과 deque를 이용해서 풀었다.
'''
from collections import deque

def bfs(p, idx):
    q = deque([idx])
    visited = [[False]*5 for _ in range(5)]
    dic = {0: [0, -1], 1:[-1, 0], 2:[0, 1], 3:[1, 0]}
    
    while q:
        x, y, d = q.popleft()
        visited[x][y] = True
        
        for i in range(4):
            nx = x + dic[i][0]
            ny = y + dic[i][1]
            nd = d + 1

            if 0 <= nx < 5 and 0 <= ny < 5 and not visited[nx][ny]:
                visited[nx][ny] = True
                
                if p[nx][ny] == 'P':
                    if nd <= 2:
                        return False

                elif p[nx][ny] == 'O':
                    if nd == 1:
                        q.append([nx, ny, nd])

    return True
                    
def solution(places):
    answer = []
    
    for p in places:
        flag = 1
        
        for i in range(5):
            for j in range(5):
                if p[i][j] == 'P':
                    result = bfs(p, [i, j, 0])
                    if not result:
                        flag = 0
                        
        answer.append(flag)
```
