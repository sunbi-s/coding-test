```python
'''
예시를 읽어서 괜히 햇갈리기만 했다.
구현문제는 말그대로 구현에서 시키것만 그대로 진행하면 답이나온다.

수학으로 하면 빈칸뚫려있는 증명문제와 비슷...

알고리즘 자체를 고민할 필요는 없다. 시키는대로 하면됨

문제에 괄호 뒤집는걸 문자열 뒤집는 걸로 생각해서 답이 틀렸다.. 독해의 문제다.

덕분에 얻은 것도 잇었는데 string[0:-1:-1] 하면 
0~ 뒤에서 2번째까지 쭉 읽고 그걸 반대로 내뱉을줄 알았는데 아니다.
뒤의 순서가 반대면 앞에 부르는 순서도 바꿔줘야됨...

즉 sring[-1::-1]로 해야 뒤에서 부터 차례대로 나온다...

+ dict 사용법에는 조금 익숙해 진거 같다.
'''
from collections import deque
def solution(p):
    p_que = deque(p)
    check = 0
    u = ''
    v = ''
    dic = {'(':1, ')':-1}
    reverse = {'(':')',')':'('}
    isCorrect = True
    
    # 1
    if p == '':
        return p
    # 2
    # u와 v로 분리
    while True:
        item = p_que.popleft()
        check += dic[item] #더해줌
        u +=item
        if check < 0:
            isCorrect = False
        if check == 0:
            break
    while p_que:
        v +=p_que.popleft()
    # 3-1
    if isCorrect:
        return u + solution(v)
    # 4-1
    new = '('
    # 4-2
    new += solution(v)
    # 4-3
    new += ')'
    # 4-4
    for u_item in u[1:-1]: 
        new += reverse[u_item]
    return new
```
