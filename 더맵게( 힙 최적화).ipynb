```python
'''
순차적 탐색 후 값 추가에서 배열은 최악의
시간복잡도를 가진다. O(n^2)이다.
이걸 heap을 쓰면 O(logn)으로 줄여진다.
배열은 조회가 O(1)인대신 append를 제외한 추가가 O(n)이다 

heapq 모듈에은 파이썬의 보통 리스트를 마치 최소 힙처럼 다룰 수 있다. 
리스트와 별개의 자료구조가 아니다!

heapify 는 list.sort()와 동일한 효과를 가진다.
파이썬에서는 heapq 모듈을 통해서 원소를 추가하거나 삭제한 리스트는 최소 힙을 유지한다.

heapify 는 O(n)이라 사실 sort O(nlogn) 대신에 그냥 heapify를 해도된다(오히려빠르다).
'''
#작동코드
import heapq

def solution(scoville, K):
    answer = 0
    scoville.sort() # heapq.heapify(scoville)
    while scoville[0] < K:
        if len(scoville) <= 1:
            return -1
        else:
            heapq.heappush(scoville, (heapq.heappop(scoville) + (heapq.heappop(scoville) * 2)))
            answer += 1
    return answer
```

```python
#내가 작성한 코드
def solution(sc, K):
    count = 0
    flag = False
    sc.sort(reverse = True)        
    
    while sc and sc[-1] < K:
        
        if len(sc) == 1:
            if flag:    
                count += 1
            else:
                count = -1
            break
        
        count += 1
        new = sc.pop()+sc.pop()*2
        
        if new < K:
            if not sc:
                if flag:
                    count += 1
                else: 
                    count = -1
                break
            
            for i in range(len(sc)): #여기가 문제다 탐색에 O(n) sort해놓은 장점을 전혀못봤다. logn으로 줄일수는 있다.
                if new <= sc[len(sc)-i-1]:
                    sc.insert(len(sc)-i,new) #삽입에 O(n) 이 걸린다. 배열의 특성의 문제라 답이없다. 못줄인다. 
                    break
                elif i == len(sc)-1:
                    sc.insert(0,new)
        else:
            flag = True
    
    if not flag:
        count = -1
    
    answer = count
    return answer
```
